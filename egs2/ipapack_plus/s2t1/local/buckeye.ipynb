```python
import os
from tqdm import tqdm
# Define the buckeyeUnicodeIPArelation and the buckeyeToUnicodeIPA function
buckeyeUnicodeIPArelation = {
    ('a', 'ʌ'),
    ('aa', 'ɑ'),
    ('aan', 'ɑ̃'),
    ('ae', 'æ'),
    ('aen', 'æ̃'),
    ('ah', 'ʌ'),
    ('ahn', 'ʌ̃'),
    ('an', 'ʌ̃'),
    ('ao', 'ɔ'),
    ('aon', 'ɔ̃'),
    ('aw', 'aʊ'),
    ('awn', 'aʊ'),
    ('ay', 'aɪ'),
    ('ayn', 'aɪ'),
    ('b', 'b'),
    ('ch', 'tʃ'),
    ('d', 'd'),
    ('dh', 'ð'),
    ('dx', 'ɾ'),
    ('eh', 'ɛ'),
    ('ehn', 'ɛ̃'),
    ('el', 'l̩'),
    ('em', 'm̩'),
    ('en', 'n̩'),
    ('eng', 'ŋ̩'),
    ('er', 'ə˞'),
    ('ern', 'ə˞'),
    ('ey', 'eɪ'),
    ('eyn', 'eɪ̃'),
    ('f', 'f'),
    ('g', 'g'),
    ('h', 'h'),
    ('hh', 'h'),
    ('hhn', 'h̃'),
    ('i', 'ɪ'),
    ('id', 'ɪ'),
    ('ih', 'ɪ'),
    ('ihn', 'ɪ̃'),
    ('iy', 'i'),
    ('iyih', 'i'),
    ('iyn', 'ɪ̃'),
    ('jh', 'dʒ'),
    ('k', 'k'),
    ('l', 'l'),
    ('m', 'm'),
    ('n', 'n'),
    ('ng', 'ŋ'),
    ('nx', 'ɾ̃'),
    ('ow', 'oʊ'),
    ('own', 'oʊ'),
    ('oy', 'ɔɪ'),
    ('oyn', 'ɔɪ'),
    ('p', 'p'),
    ('q', 'ʔ'),
    ('r', 'ɹ'),
    ('s', 's'),
    ('sh', 'ʃ'),
    ('t', 't'),
    ('th', 'θ'),
    ('tq', 'ʔ'),
    ('uh', 'ʊ'),
    ('uhn', 'ʊ̃'),
    ('uw', 'u'),
    ('uwix', 'u'),
    ('uwn', 'ũ'),
    ('v', 'v'),
    ('w', 'w'),
    ('y', 'j'),
    ('z', 'z'),
    ('zh', 'ʒ')
}

def buckeyeToUnicodeIPA(buckeyeSymbol):
    mapping = dict(buckeyeUnicodeIPArelation)
    return mapping.get(buckeyeSymbol, '')  # Return empty string if not found

# Define a function to process the .phn file and convert it to IPA
def convert_phn_to_ipa(filename):
    ipa_transcription = []
    
    # Read the file content
    with open(filename, 'r') as file:
        for line in file:
            parts = line.split()  # Split the line by spaces
            if len(parts) > 1:
                phonetic_symbol = parts[2]  # Extract the phonetic symbol
                ipa_symbol = buckeyeToUnicodeIPA(phonetic_symbol)  # Convert to IPA
                ipa_transcription.append(ipa_symbol)  # Add to the list
    
    return ''.join(ipa_transcription)  # Join the IPA symbols into a single string

# Function to save the IPA result to a file
def save_ipa_to_file(input_filename, ipa_result):
    # Create a new filename by replacing the extension with '.ipa'
    base_filename, _ = os.path.splitext(input_filename)
    output_filename = f"{base_filename}.ipa"
    
    # Write the IPA result to the output file
    with open(output_filename, 'w') as output_file:
        output_file.write(ipa_result)
    

# Function to process all .phn files in nested directories
def process_all_phn_files_in_directory(root_dir):
    for dirpath, _, filenames in os.walk(root_dir):
        for filename in tqdm(filenames):
            if filename.endswith('.phn'):  # Check if the file is a .phn file
                full_file_path = os.path.join(dirpath, filename)
                ipa_result = convert_phn_to_ipa(full_file_path)
                save_ipa_to_file(full_file_path, ipa_result)

# Test the function by providing the root directory to process all .phn files
root_directory = '/scratch/lingjzhu_root/lingjzhu1/lingjzhu/buckeye_split'  # Replace with your root directory
process_all_phn_files_in_directory(root_directory)
```
